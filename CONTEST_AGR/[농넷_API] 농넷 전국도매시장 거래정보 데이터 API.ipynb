```python
import pandas as pd
import urllib.request
import json
import datetime
```

### 어제의 농넷 데이터 불러오기

시간별, 도매시장별로 데이터가 계속 추가 업데이트 됩니다.  
전체 데이터를 얻기 위해서 하루 전날(D-1)을 선택.

```python
def getdata(date):

    # today = datetime.datetime.today()
    # yesterday = (today - datetime.timedelta(1)).strftime('%Y%m%d') 
    url = f'https://www.nongnet.or.kr/api/whlslDstrQr.do?sdate={date}' # sdate = 날짜

    response = urllib.request.urlopen(url).read()
    response = json.loads(response)

    data = pd.DataFrame(response['data'])
    data.to_csv(f'/home/lab10/real_con/AT_TSALET_ALL_{date}.csv')
    print(data)
    return data
```
